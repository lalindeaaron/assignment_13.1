```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 
```

```python
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

Setting up database:

```python
sqladb = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
```

```python
sqladb.columns=['age','workclass','fnlwgt','education','educationNum','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
```

Question 1: Select 10 records from the adult sqladb

```python
pysqldf("SELECT * FROM sqladb LIMIT 10;")
```

Question 2: Show me the average hours per week of all men who are working in private sector

```python
pysqldf("""select avg("hours-per-week") from sqladb where sex=(" Male") and workclass=(" Private");""")
```

Question 3: Show me the frequency table for education, occupation and relationship, separately

```python
pysqldf("""select education,count(*) from sqladb GROUP BY("education");""")
```

```python
pysqldf("""select occupation,count(*) from sqladb GROUP BY("occupation");""")
```

```python
pysqldf("""select relationship,count(*) from sqladb GROUP BY("relationship");""")
```

Question 4: Are there any people who are married, working in private sector and having a masters
degree?

```python
pysqldf("""select * from sqladb where education=" Masters" and workclass=(" Private") and ("marital-status") Not like "%Never-married%";""")
```

Question 5: What is the average, minimum and maximum age group for people working in
different sectors

```python
pysqldf("""select workclass,avg(age),min(age),max(age) from sqladb GROUP BY("workclass")""")
```

Question 6: Calculate age distribution by country

```python
pysqldf(""" SELECT ("native-country"),age, COUNT(age) AS Frequency FROM sqladb GROUP BY "age","native-country" ORDER BY COUNT(age) DESC""")
```

Question 7: Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and
'capital-loss'

```python
pysqldf("""select ("capital-gain"-"capital-loss") AS "Net-Capital-Gain" from sqladb;""")
```
